```python
def sapa (nama="Mahasiswa") :
    print(f"Halo,, {nama}!")

sapa()        #tanpa argumen
sapa("Andi")  #dengan argumen
```

Praktikum 04-FUngsi dengan *args dan **kwargs

```python
def jumlahkan(*args) :
    return sum(args)

def biodata(**kwargs) :
    for key, value in kwargs.items() :
        print(f"{key}: {value}")

print("Total:", jumlahkan(1,2,3,4,5))
biodata(nama="Budi", umur=18, Prodi="TRK")
```

```python
x = 10

def ubah () :
    x = 5
    print("Nilai lokal:", x)

ubah()
print("Nilai global:", x)
```

Praktikum 07-Fungsi Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

Penugasan 4-Fungsi dengan **args**

```python
def jumlahkan_semua(*angka):
    total = sum(angka)
    return total


hasil_3 = jumlahkan_semua(10, 20, 30)
hasil_5 = jumlahkan_semua(1, 2, 3, 4, 5)
hasil_7 = jumlahkan_semua(5, 5, 5, 5, 5, 5, 10)


print(f"Total 3 bilangan: {hasil_3}")
print(f"Total 5 bilangan: {hasil_5}")
print(f"Total 7 bilangan: {hasil_7}")
```

Penugasan 5-Fungsi dengan **kwargs**

```python
def info_barang(**data):
    print("--- Informasi Barang ---")
    for kunci, nilai in data.items():
        print(f"{kunci}: {nilai}")

info_barang(Nama="Buku", Harga=15000, Stok=30)

print("\nContoh tambahan:")
info_barang(Nama="Laptop", Merek="Asus", Garansi="2 Tahun", Harga=8000000)
```

Penugasan 7-Fungsi Lambda

```python
kuadrat = lambda bilangan: bilangan ** 2

keliling_lingkaran = lambda r: 2 * 3.14 * r

print(f"Kuadrat dari 5: {kuadrat(5)}")
print(f"Keliling lingkaran (r=10): {keliling_lingkaran(10)}")
```

Penugasan 8-Fungsi Decorator

```python
def logger(fungsi):
    def wrapper():
        print("Memulai fungsi...")
        fungsi()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, ini adalah fungsi sapaan sederhana.")

halo()
```

Penugasan 9-Studi Kasus Nyata-Program Nilai

```python
def input_nilai():
    return [70, 85, 90, 60]

def hitung_rata(data):
    return sum(data) / len(data)

def status_kelulusan(rata):
    if rata >= 60:
        return "LULUS"
    else:
        return "TIDAK LULUS"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

daftar_nilai = input_nilai()
rata_rata = hitung_rata(daftar_nilai)
hasil_akhir = nilai_akhir(80, 75)

print(f"Daftar Nilai: {daftar_nilai}")
print(f"Rata-rata: {rata_rata}")
print(f"Status: {status_kelulusan(rata_rata)}")
print(f"Nilai Akhir Berbobot: {hasil_akhir}")
```
